```python
import pandas as pd
```

```python
business_data = pd.read_csv("business.csv")
review_data = pd.read_csv("review.csv")
friend_data = pd.read_csv("user.csv")
```

```python
review_data['reviews'] = review_data.text
```

```python
review_data.head()
```

```python
review = review_data[['user_id', 'business_id', 'stars', 'reviews']]
review['review_star'] = review.stars
review = review[['user_id', 'business_id', 'reviews', 'review_star']]
review
```

```python
friends = friend_data[['user_id', 'friends']]
```

```python
business_data['restaurant_star'] = business_data.stars
business_data
```

```python
business_data = business_data[['business_id', 'categories', 'city', 
                            'is_open', 'latitude', 'longitude', 'name', 
                            'review_count', 'restaurant_star']]
business_data
```

```python
df = pd.merge(business_data, review,on=['business_id'], how='inner')
```

```python
data_Charlotte = df[df.city == 'Charlotte']
```

```python
data = pd.merge(data_Charlotte,friends,on=['user_id'], how='inner')
```

```python
import numpy as np
```

```python
data.head()
```

```python
def counter(x):
    return [len(set(filter(None, i.split(',')))) for i in x]

for col in ['friends']:
    data['count'+col] = counter(data[col])

data.sort_values(by=['countfriends'],ascending=False)

```

```python
friend_data = data[data.countfriends >= 1000]

print(len(friend_data.user_id.unique()))

print(len(friend_data.business_id.unique()))

print(len(friend_data))
```

```python
data_frame = friend_data.to_csv (r'data_friends-08-12.csv', index = None, header=True)
```

```python
friends = pd.read_csv('data_friends-08-12.csv')
friends
```

```python
uniqnames = friends['business_id'].unique() # get unique names
newdata = []                 # data list for output dataframe
for u in uniqnames:          # for each unique name
    subreview = friends[friends['business_id'] == u] # get rows with this unique name
    s = ""
    for i in subreview['reviews']:
        s += "("+i+"),"      # join all info cells for that name
    newdata.append([u, s[:-1]]) # remove trailing comma from infos & add row to data list

newdf = pd.DataFrame(data=newdata, columns=['business_id','reviews'])
print(newdf)
```

```python
data_frame = newdf.to_csv (r'business_revies_merge_08_12.csv', index = None, header=True)
```
